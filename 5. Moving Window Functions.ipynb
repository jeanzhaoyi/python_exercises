```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ts = pd.Series(np.random.randn(20), 
              pd.date_range('7/1/2016', freq = 'H', periods= 20))

ts_lagged = ts.shift(-5) # forward or backward shift by the sign

plt.plot(ts, color = 'blue')
plt.plot(ts_lagged, color = 'red')
```

#### Window functions are like aggregation functions
<p>You can use them in conjunction with .resample() which will soon be a deferred operation

```python
df = pd.DataFrame(np.random.randn(600, 3), 
                  index = pd.date_range('7/1/2016', freq = 'S', periods = 600),
                  columns = ['A', 'B', 'C'])
```

```python
#pd.rolling_mean(df, window = 2)[1:10] # in future versions you want to resample separately
r = df.rolling(window = 10)
#r.agg, r.apply, r.count, r.exclusions, r.max, r.median, r.name, r.quantile, r.kurt, r.cov, r.corr, r.aggregate, r.std, r.skew, r.sum, r.var
df.plot(style = 'k--')
r.mean().plot(style = 'k')
```

```python
r = df.rolling(window=10)
df['A'].plot(color = 'gray')
r.mean()['A'].plot(color = 'red')
```

```python
r.quantile(0.5).plot()
```

```python
r.agg(['sum','var'])
```

```python
#exponentially weighted
pd.ewma(df, span = 15, min_periods = 2)[1:10]
```

```python
# can also easily plot a rolling average per column
df = pd.DataFrame(np.random.randn(1000, 4), index = pd.date_range('6/6/16', periods = 1000), columns = ['A', 'B', 'C', 'D'])

df.head()
```

```python
df = df.cumsum()
df.rolling(window = 50).sum().plot(subplots=True)
```

```python
# What about a custom function?
df.rolling(window = 10, center = False).apply(lambda x: np.fabs(x - x.mean()).mean())
```

```python
df.rolling(window= 10, center=False).apply(lambda x: x[1]/x[2])[10:22]
```

```python
df.rolling(window= 10, center=False).apply(lambda x: x[0] - x[-1])[10:22]
# difference from the first to the last in the window
```

```python
# Expanding windows
# Yields the value of the statistic with all the data available up to that point in time
df.expanding(min_periods = 1).mean()[1:5]
```

### Get monthly data from daily data using rolling window?

```python
ts_long = pd.Series(np.random.randn(200),
                   pd.date_range('7/1/2016', freq = 'D', periods = 200))

ts_long.resample('M').mean().rolling(window = 3).mean().plot() 
# 3 months at a time rolling window
```

### Expanding windows 

```python
df.expanding(min_periods= 1).mean().plot()
```

```python
ts = pd.Series(np.random.randn(1000), 
              index = pd.date_range(start = '1/1/16', freq = 'D', periods=1e3))
ts.ewm(span = 60, freq = 'D', min_periods = 0, adjust = True).mean().plot()
# exponentially weighted moving average
ts.rolling(window = 60).mean().plot()
```

```python
r = ts.rolling(window=20)

r.apply(lambda x: sorted(x)[round(len(x)*.5)])  
# get position of the series that is at the 50th percentile, so sorted first
# same as quantile(.5)
```

```python
r.agg(['sum','count']).head() # more than one function at the same time
# can include custom function too
```
